### Exercise 03.1

```python
prefix = "exercise03-1/"
```

In this exercise I compute an optimal price for a call or put option, sampling directly the asset price at the expiry as a geometric Bronwnian motion.
The dotted red lines mark the prices obtained from the analytical solution, from Black-Scholes theory.

```python
call_avg, call_std = np.loadtxt(prefix + "call-direct-price.dat", unpack = True)
plt.figure(figsize = (15,10))
plt.hlines(14.975790778311286, 0, len(call_avg), colors = "red", linestyle = "dotted")
plt.errorbar(range(1, len(call_avg) + 1), call_avg, yerr = call_std)
plt.title("Call option")
plt.xlabel("Number of blocks")
plt.ylabel("Option price")
plt.grid(True)
plt.show()
```

```python
call_avg, call_std = np.loadtxt(prefix + "put-direct-price.dat", unpack = True)
plt.figure(figsize = (15,10))
plt.hlines(5.4595325819072364, 0, len(call_avg), colors = "red", linestyle = "dotted")
plt.errorbar(range(1, len(call_avg) + 1), call_avg, yerr = call_std)
plt.title("Put option")
plt.xlabel("Number of blocks")
plt.ylabel("Option price")
plt.grid(True)
plt.show()
```
