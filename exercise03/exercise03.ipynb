```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import string
from scipy.optimize import curve_fit
```

### Exercise 03.1

In this exercise I compute an optimal price for a call or put option, sampling directly the asset price at the expiry as a geometric Bronwnian motion.
The dotted red lines mark the prices obtained from the analytical solution, from Black-Scholes theory.

```python
call_avg, call_std = np.loadtxt("call_direct_price.dat", unpack = True)
put_avg, put_std = np.loadtxt("put_direct_price.dat", unpack = True)

fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (12, 5))
axes = axes.flatten()

# Call option price plot
axes[0].errorbar(
    range(1, len(call_avg) + 1),
    call_avg,
    yerr = call_std,
    elinewidth = 1,
    label = "Simulation"
)
axes[0].hlines(
    14.975790778311286,
    0,
    len(call_avg),
    colors = "red",
    linestyle = "dotted",
    label = "Analytical solution"
)
axes[0].set_title("Call option")
axes[0].set_xlabel("Number of blocks")
axes[0].set_ylabel("Option price")
axes[0].legend()
axes[0].grid(True)

# Put option price plot
axes[1].errorbar(
    range(1, len(put_avg) + 1),
    put_avg,
    yerr = put_std,
    elinewidth = 1,
    label = "Simulation"
)
axes[1].hlines(
    5.4595325819072364,
    0,
    len(put_avg),
    colors = "red",
    linestyle = "dotted",
    label = "Analytical solution"
)
axes[1].set_title("Put option")
axes[1].set_xlabel("Number of blocks")
axes[1].set_ylabel("Option price")
axes[1].legend()
axes[1].grid(True)

fig.tight_layout()
plt.show()
```
