```python
gen, best_fit_length = np.loadtxt("best_fit_length.dat", unpack = True)
gen, avg_best_length = np.loadtxt("averaged_best_length.dat", unpack = True)
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (12, 5))
axes = axes.flatten()

axes[0].plot(best_fit_length)
axes[0].set_title("Length of the best path")
axes[0].set_xlabel("Generation")
axes[0].set_ylabel("Length")
axes[0].grid(True)

axes[1].plot(avg_best_length)
axes[1].set_title("Average length of the best half of the population")
axes[1].set_xlabel("Generation")
axes[1].set_ylabel("Length")
axes[1].grid(True)

plt.show()
```
