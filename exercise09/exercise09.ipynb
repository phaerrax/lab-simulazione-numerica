### Exercise 09.1

##### Cities on a circle
With a genetic algorithm I calculated the solution to the "travelling salesman problem" for a set of 30 cities randomly distributed on a circle. To the right of the plots, the solution is represented graphically. 
The data can be generated by running `./exercise09 circle`, making sure to create a directory named `circle` beforehand (the results will be stored in there).

```python
prefix = "circle/"

gen, best_fit_length = np.loadtxt(prefix + "best_fit_length.dat", unpack = True)
gen, avg_best_length = np.loadtxt(prefix + "averaged_best_length.dat", unpack = True)
fig, axes = plt.subplots(nrows = 1, ncols = 3, figsize = (15, 4.5))
axes = axes.flatten()

axes[0].plot(best_fit_length)
axes[0].set_title("Length of the best path")
axes[0].set_xlabel("Generation")
axes[0].set_ylabel("Length")
axes[0].grid(True)

axes[1].plot(avg_best_length)
axes[1].set_title("Average length of the best half of the population")
axes[1].set_xlabel("Generation")
axes[1].set_ylabel("Length")
axes[1].grid(True)

with open(prefix + "evolution.dat") as file:
    for line in file:
        pass
    solution = [int(n) for n in line.split()[1:]]
    # Remove the first number: it is the
    # number of the current generation.

cities = []
best_path = []

with open(prefix + "cities.dat") as file:
    for line in file:
        cities.append([float(c) for c in line.split()])

for n in solution:
    best_path.append(cities[n])

for p in range(len(best_path)):
    X = (best_path[p - 1][0], best_path[p][0])
    Y = (best_path[p - 1][1], best_path[p][1])
    axes[2].plot(X, Y, "ro-")
axes[2].axis("off")

plt.show()
```
